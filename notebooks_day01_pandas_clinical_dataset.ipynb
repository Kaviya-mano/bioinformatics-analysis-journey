```python
# Bioinformatics Analysis Journey
# Day 01 – Clinical Dataset Exploration with Pandas
# Author: Kaviya Mano
import pandas as pd

data = {
"Patient_ID":["P1","P2","P3","P4","P5"],
"Age":[45,50,38,60,55],
"Tumor_Size_cm":[3.2,2.1,4.5,3.8,2.9],
"Mutation":["TP53","EGFR","KRAS","TP53","EGFR"],
"Survival_Months":[24,36,12,18,30]
}

df = pd.DataFrame(data)

print(df)
df.head()
print(df.shape)
print(df.columns)
print(df.info())
df.describe()
df["Mutation"].value_counts()
print(df["Age"].mean())
print(df["Tumor_Size_cm"].mean())
print(df.groupby("Mutation")["Survival_Months"].mean())
```
